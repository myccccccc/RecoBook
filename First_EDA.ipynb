```python
import numpy as np
import pandas as pd
import csv
```

```python
BOOK_NUM = 8000
USER_NUM = 15000
```

```python
ratings = pd.read_csv("UCSD_Book_Graph/ratings.csv")
print(len(ratings))
ratings.head()
```

```python
useronly = ratings[["user_id", "rating"]].groupby("user_id").agg(len).sort_values(by="rating",ascending = False).head(USER_NUM).reset_index()
useronlyratings = ratings[ratings["user_id"].isin(useronly["user_id"])]
bookonly = useronlyratings[["book_id", "rating"]].groupby("book_id").agg(len).sort_values(by="rating",ascending = False).head(BOOK_NUM).reset_index()
finalratings = useronlyratings[useronlyratings["book_id"].isin(bookonly["book_id"])]
print(len(finalratings))
finalratings.head()
```

```python
print(len(finalratings["user_id"].unique()))
print(len(finalratings["book_id"].unique()))
```

```python
finalratings = finalratings.drop(columns=["Unnamed: 0"])
finalratings.head()
```

```python
books = pd.read_csv("UCSD_Book_Graph/books.csv", quoting=csv.QUOTE_NONE)
print(len(books))
books.head()
```

```python
books = books[books["book_id"].isin(bookonly["book_id"])]
books = books.reset_index(drop=True)
books["newbook_id"] = books.index + 1
finalbooks = books
print(len(finalbooks))
finalbooks.head()
```

```python
finalratings = finalratings.merge(finalbooks[['book_id','newbook_id']], how='left', on= ['book_id'])
print(len(finalratings))
finalratings.head()
```

```python
finalratings['newuser_id'] = finalratings.groupby('user_id').grouper.group_info[0]+1
print(len(finalratings))
finalratings.head()
```

```python
finalratings = finalratings.drop(['user_id','book_id'], axis=1)
print(len(finalratings))
finalratings.head()
```

```python
finalratings.to_csv("finalratings.csv", encoding='utf8')
```

```python
finalbooks = finalbooks.drop(['book_id'], axis=1)
print(len(finalbooks))
finalbooks.head()
```

```python
authors = pd.read_csv("UCSD_Book_Graph/authors.csv", index_col="author_id")
authors.head()
```

```python
def author_ids2author_names(author_ids):
    author_ids = str(author_ids)
    if author_ids == "|":
        return None
    author_names = ""
    for i in author_ids.split("|")[:-1]:
        author_names += authors.loc[int(i), "name"]
        author_names += "|"
    return author_names[:-1]
```

```python
finalbooks["author_ids"] = finalbooks["author_ids"].apply(author_ids2author_names)
print(len(finalbooks))
finalbooks.head()
```

```python
finalbooks.info()
```

```python
finalbooks["publication_year"] = finalbooks["publication_year"].fillna(finalbooks["publication_year"].mean())
finalbooks["language_code"] = finalbooks["language_code"].fillna("Unknown")
finalbooks["description"] = finalbooks["description"].fillna("Unknown")
finalbooks["num_pages"] = finalbooks["num_pages"].fillna(finalbooks["num_pages"].mean())
finalbooks.head()
```

```python
finalbooks.info()
```

```python
finalbooks = finalbooks.rename(columns={"author_ids": "author_names"})
finalbooks.head()
```

```python
finalbooks.to_csv("finalbooks.csv", encoding="utf8")
```
